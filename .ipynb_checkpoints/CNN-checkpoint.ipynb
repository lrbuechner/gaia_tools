## Positive Clusters

...$get\_clusters(g, n, N, d, D)$ $\mapsto$ [ 'clustername', pandas df of cluster members]
 
 - $g$ = patition sphere scale [1, inf]. The partition k=1 is the sphere with recorded cluster radius. k=2 is double that
 - $n$ = subset open cluster catalogue to look for star clusters w minum $n$ stars
 - $N$ = hard star count threshold in case the query doesnt reflect $n$. Sometimes the clusters arent particularly spherical

 - $d,D$ = min and max distances in parsecs (pc)
 
###### Below is a quick demo querying nearby open clusters as well as plots showing the type of data collected.

I use a Gaussian Mixture Model to seperate the cluster members from field stars. It's necessary to remove this excess noise in order to collect meaningful training data.

## Negative Clusters

...$get\_fieldstars(i, n, N, d, D)$ $\mapsto$ [ 'clustername', pandas df of cluster members]
 
 - $i$ = how many random spherical partitions of stars with at least $randint(n,N)$ stars 
 - $n,N$ = it's desirable to randomly select a different integer number of stars within the range for each iteration [$n,N$]
 - $d,D$ = min and max distances in parsecs (pc)
 
###### This data will be added to the known clusters and used as a training set for a CNN.

# CNN Processing/Training

- CNN from scratch: https://victorzhou.com/blog/intro-to-cnns-part-2/
- Official Keras: https://keras.io/examples/mnist_cnn/

```python
from sklearn import preprocessing
import matplotlib.pyplot as plt

def prep_data(df, boolie):
    
    if boolie == 'test':
        g, rp, bp, color = viz(df).get_CMD()
        CMD_data = viz(df).get_CMD()
        # in viz class there is normalize method that is use for making heatmap
        norm_g, norm_color = viz(df).normalize(CMD_data[0],CMD_data[3])
        pixels = plt.hist2d(
            norm_color,norm_g, 
            bins=[np.linspace(-3,3,100),np.linspace(-10,10,100)])[0]
        plt.close() # prevents plt from showing
        norm_pix = preprocessing.normalize(pixels)
        return norm_pix
    
    else:
        # takes advantages of plt.hist2d to convert normalized 
        # scatter plot into matrix of values (intensity), units are (color, absolute mag)
        images = []
        ids = df['name'].unique()
        for n in range(len(ids)):
            subset = df[df['name'] == ids[n]]
            g, rp, bp, color = viz(subset).get_CMD()
            CMD_data = viz(subset).get_CMD()
            # in viz class there is normalize method that is use for making heatmap
            norm_g, norm_color = viz(subset).normalize(CMD_data[0],CMD_data[3])
            pixels = plt.hist2d(
                norm_color,norm_g, 
                bins=[np.linspace(-3,3,100),np.linspace(-10,10,100)])[0]
            plt.close() # prevents plt from showing
            norm_pix = preprocessing.normalize(pixels)
            images.append(norm_pix)

        # generates corresponding labels depending on input type {0,1} (binary classification)
        if boolie == True:
            return images, np.array([1 for i in range(len(images))])
        elif boolie == False:
            return images, np.array([0 for i in range(len(images))])
```

## Consolidating and Prepping Data

$prep\_data(df, boolie)$ $\mapsto$ normalized grid of pixels, labels $\in$ {0,1}

- $df$ = either a dataframe of clusters or field stars, not both
- $boolie$ = True for df of clusters | False for df of field stars

```python
x_true, y_true = prep_data(true, True)
x_false, y_false = prep_data(false, False)

# consolidating everything 
X = np.array(list(x_true)+list(x_false))
Y = np.append(y_true,y_false)
```

### Importing necessary modules 

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from keras.layers.normalization import BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from sklearn.metrics import confusion_matrix

batch_size = 32
num_classes = 2
epochs = 100
input_shape = (99, 99, 1)
```

```python
# splitting data for training and model validation

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size = 0.1, random_state=77)
```

```python
# reshaping for keras

X_train = X_train.reshape(X_train.shape[0], 99, 99,1)
X_val = X_val.reshape(X_val.shape[0], 99, 99,1)

Y_train = to_categorical(Y_train)
Y_val = to_categorical(Y_val)
```

```python
# standard 2D CNN - By no means does the network have optimal architecture  

CNN = Sequential()
CNN.add(Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_initializer='he_normal', input_shape=input_shape))
CNN.add(Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_initializer='he_normal'))
CNN.add(MaxPool2D((9, 9)))
CNN.add(Dropout(0.20))
CNN.add(Conv2D(128, (9, 9), activation='relu', padding='same', kernel_initializer='he_normal'))
CNN.add(Conv2D(128, (9, 9), activation='relu', padding='same', kernel_initializer='he_normal'))
CNN.add(MaxPool2D((2, 2)))
CNN.add(Dropout(0.25))
CNN.add(Conv2D(64, (3, 3), activation='relu', padding='same', kernel_initializer='he_normal'))
CNN.add(Dropout(0.25))
CNN.add(Flatten())
CNN.add(Dense(32, activation='relu'))
CNN.add(BatchNormalization())
CNN.add(Dropout(0.25))
CNN.add(Dense(num_classes, activation='softmax'))

CNN.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.RMSprop(),
              metrics=['accuracy'])

extra_params = ImageDataGenerator(
        featurewise_center=False,  
        samplewise_center=False, 
        featurewise_std_normalization=False, 
        samplewise_std_normalization=False,  
        zca_whitening=False,  
        rotation_range=15, # 
        zoom_range = 0.1, 
        width_shift_range=0.1,  
        height_shift_range=0.1,  
        horizontal_flip=False,  
        vertical_flip=False)  

LR_controls = ReduceLROnPlateau(monitor='val_acc', patience=3, verbose=1, factor=0.5, min_lr=0.0001)
```

```python
extra_params.fit(X_train)
run = CNN.fit_generator(extra_params.flow(X_train,Y_train, batch_size=batch_size),
          epochs = epochs, validation_data = (X_val,Y_val),
          verbose = 1, steps_per_epoch=X_train.shape[0] // batch_size
          , callbacks=[LR_controls])
```

## Training Results

```python
accuracy = run.history['acc']
val_accuracy = run.history['val_acc']
loss = run.history['loss']
val_loss = run.history['val_loss']
epochs = range(len(accuracy))

plt.figure(figsize = (8,5))
plt.plot(epochs, accuracy, 'r', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'k', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.show()

plt.figure(figsize = (8,5))
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'k', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

## Potential Issues

Notice that the model has a tendecy to predict false positives. This is likely due to two reasons. Recall that the way I generated my labeled data was with two functions, one to collect random field stars in a given region of space or to query known star cluster. 

1) When collecting field stars, the method randomly selects a spherical partition in space and expands it until it contains a specified number of stars. Sometimes it will randomly select a region in space in which the chemical composition of stars is similar to that of a star cluster, this is characteristic of tidally stripped open clusters that are still relatively young. An example of this is the heatmap on the bottom right. It might be an interesting to further inspect these regions in order to see if there are any interesting interatctions with neighboring stars. 

2) Some of the more distance star clusters >4kpc away are hard to resolve due to high parallax error. This results in only being able to recover a subset of the true cluster members. Few stars translates to little information from which the CNN can extract structure from -> bad predictions.

## Looking Forward

There are several methods that can be used to identify the astrometric clustering expressed by open clusters in the local galaxy. Below is just one of those examples, DBSCAN. https://en.wikipedia.org/wiki/DBSCAN

```python
val_df = pd.read_csv(r'\Users\Landon\Documents\Jupyter\Astronomy\Moving Group Detection\Data\validateZ.csv')
val_df['dist_pc'] = 1000/val_df['parallax']
val_df['PM'] = np.sqrt(val_df['pmra']**2 + val_df['pmdec']**2)
```

```python
from sklearn.cluster import DBSCAN

# detecting clusters with DBSCAN
value_matrix = val_df[['ra','dec','pmra','pmdec']]
clustering = DBSCAN(eps=.15, min_samples=30).fit_predict(value_matrix)
val_df['DBSCAN_label'] = clustering
```
